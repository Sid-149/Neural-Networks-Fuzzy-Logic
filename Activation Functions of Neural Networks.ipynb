# Different Activation Function of Neural Networks

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Identity Funtion

```python
def identity_func(x):
    y = x
    return y
```

```python
x = np.linspace(0.1, 20, 50) 
print(x)
y = identity_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Identity Function')
```

## Binary Step Function 

```python
def binary_step_func(x):
    y = list()
    for x1 in x: 
        if x1 >= 0:
            y.append(1)
        else: y.append(0)
    return y
```

```python
x = np.linspace(-20, 20, 50) 
print(x)
y = binary_step_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Binary Step Function')
```

## Bipolar Step Function 

```python
def bipolar_step_func(x):
    y = list()
    for x1 in x: 
        if x1 >= 0:
            y.append(1)
        else: y.append(-1)
    return y
```

```python
x = np.linspace(-20, 20, 50) 
print(x)
y = bipolar_step_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Bipolar Step Function')
```

## Binary Sigmoidal Function 

```python
def binary_sigmoid_func(x):
    y = 1/(1 + np.exp(-lambda1 * x))
    return y
```

```python
x = np.linspace(-20, 20, 50) 
print(x)
lambda1 = 2
y = binary_sigmoid_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Binary Sigmoidal Function')
```

## Bipolar Sigmoidal Function

```python
def bipolar_sigmoid_func(x):
    y = (1 - np.exp(-lambda1 * x))/(1 + np.exp(-lambda1 * x))
    return y
```

```python
x = np.linspace(-20, 20, 50) 
print(x)
lambda1 = 2
y = bipolar_sigmoid_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Bipolar Sigmoidal Function')
```

## Ramp Function 

```python
def ramp_func(x):
    y = list()
    for x1 in x: 
        if x1 > 1:
            y.append(1)
        elif 0 <= x1 <= 1:
            y.append(x1)
        else: y.append(0)
    return y
```

```python
x = np.linspace(-20, 20, 50) 
print(x)
y = ramp_func(x)
print(y)
```

```python
#plotting graph
plt.plot(x,y)
plt.xlabel('Net Input')
plt.ylabel('Output')
plt.title('Ramp Function')
```
